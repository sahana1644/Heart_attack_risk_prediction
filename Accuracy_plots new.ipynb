```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xgboost_accuracies=[0.975,0.8913043478260869,0.8333333333333334,0.5738271604938271]
```

```python
random_forest_accuracies=[0.97,0.8967391304347826,0.8518518518518519,0.7467695473251028]
```

```python
stacking_classifier_accuracies=[0.965,0.8804347826086957,0.8518518518518519,0.5101646090534979]
```

```python
svm_accuracies = [0.96,0.8695652173913043,0.8518518518518519,0.7667695473251028]
```

```python
accuracies_grouped = [xgboost_accuracies, random_forest_accuracies, stacking_classifier_accuracies, svm_accuracies]
```

```python
classifiers = ['XGBoost', 'Random Forest', 'Stacking Classifier', 'SVM']

datasets = ['Dataset 1', 'Dataset 2', 'Dataset 3', 'Dataset 4']

num_classifiers = len(classifiers)

num_groups = len(xgboost_accuracies)

bar_width = 0.2

bar_positions = np.arange(num_groups)

plt.figure(figsize=(10, 6))

for i in range(num_classifiers):
    plt.bar(bar_positions + i * bar_width, accuracies_grouped[i], width=bar_width, label=classifiers[i])

plt.xlabel('Dataset')
plt.ylabel('Accuracy')
plt.title('Accuracies for Different Models')
plt.xticks(bar_positions + (bar_width * (num_classifiers - 1)) / 2, datasets)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Given lists
dataset_1 = [0.975, 0.97, 0.965, 0.96]
dataset_2 = [0.8913043478260869, 0.8967391304347826, 0.8804347826086957, 0.8695652173913043]
dataset_3 = [0.8333333333333334, 0.8518518518518519, 0.8518518518518519, 0.8518518518518519]
dataset_4 = [0.5738271604938271, 0.7467695473251028, 0.5101646090534979,0.7667695473251028]

num_groups = len(dataset_1)
bar_width = 0.2

x = np.arange(num_groups)

plt.figure(figsize=(10, 6))  # Adjust width and height as needed

plt.bar(x, dataset_1, width=bar_width, label='Dataset 1')
plt.bar(x + bar_width, dataset_2, width=bar_width, label='Dataset 2')
plt.bar(x + 2*bar_width, dataset_3, width=bar_width, label='Dataset 3')
plt.bar(x + 3*bar_width, dataset_4, width=bar_width, label='Dataset 4')

plt.title('Performance of Different Classifiers')
plt.xlabel('Index')
plt.ylabel('Accuracy')
plt.xticks(x + 1.5*bar_width, ['XGBoost', 'Random Forest', 'Stack Classifier', 'SVM'])
plt.legend()

plt.show()
```
